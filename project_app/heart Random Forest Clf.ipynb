```python
class rf_model():
    def __init__(self,df):
        self.df= df
    def train_test(self):
        self.x= self.df.drop('target',axis=1)
        self.y= self.df['target']
        self.columns ={'columns':list(self.x.columns)}
        self.x_train,self.x_test,self.y_train,self.y_test= train_test_split(self.x,self.y,test_size=0.2,random_state=42,stratify=self.y)

    def model_building(self):
        self.rf_model= RandomForestClassifier()
        self.rf_model.fit(self.x_train,self.y_train)
    
    def evalution_metrics_for_train_data(self):
        y_pred=self.rf_model.predict(self.x_train)
        cnf_metrics=confusion_matrix(self.y_train,y_pred)
        print('Confusion Metrics:\n',cnf_metrics)
        print('*'*80)
        clf_report=classification_report(self.y_train,y_pred)
        print('Classification Report:\n',clf_report)
        print('*'*80)
        accuracy=accuracy_score(self.y_train,y_pred)
        print('Accuracy:',accuracy)
    
    def evalution_metrics_for_test_data(self):
        y_pred_test=self.rf_model.predict(self.x_test)
        cnf_metrics=confusion_matrix(self.y_test,y_pred_test)
        print('Confusion Metrics:\n',cnf_metrics)
        print('*'*80)
        clf_report=classification_report(self.y_test,y_pred_test)
        print('Classification Report:\n',clf_report)
        print('*'*80)
        accuracy=accuracy_score(self.y_test,y_pred_test)
        print('Accuracy:',accuracy)
    
    def hyperparameter(self):
        self.train_test()
        
        self.rf_model= RandomForestClassifier()
        
        param= {  'n_estimators':np.arange(20,200),
                    'criterion':['gini','entropy'],
                    'max_depth':np.arange(3,8),
                    'min_samples_split':np.arange(2,20),
                    'min_samples_leaf':np.arange(2,20),
                    #'max_features':[3,4],
                    'bootstrap':[True],
                    'oob_score':[True]}
        self.gscv_rf=RandomizedSearchCV(self.rf_model,param,cv=5)
        self.gscv_rf.fit(self.x_train,self.y_train)
    
    def evalution_metrics_for_train_data_with_hyper(self):
        y_pred=self.gscv_rf.predict(self.x_train)
        cnf_metrics=confusion_matrix(self.y_train,y_pred)
        print('Confusion Metrics:\n',cnf_metrics)
        print('*'*80)
        clf_report=classification_report(self.y_train,y_pred)
        print('Classification Report:\n',clf_report)
        print('*'*80)
        accuracy=accuracy_score(self.y_train,y_pred)
        print('Accuracy:',accuracy)
    def evalution_metrics_for_test_data_with_hyper(self):
        y_pred_test=self.gscv_rf.predict(self.x_test)
        cnf_metrics=confusion_matrix(self.y_test,y_pred_test)
        print('Confusion Metrics:\n',cnf_metrics)
        print('*'*80)
        clf_report=classification_report(self.y_test,y_pred_test)
        print('Classification Report:\n',clf_report)
        print('*'*80)
        accuracy=accuracy_score(self.y_test,y_pred_test)
        print('Accuracy:',accuracy)
    def model_files(self):
        self.hyperparameter()
        with open ('RF_model.pkl','wb') as f:
            pickle.dump(self.gscv_rf,f)
        
        with open('project_data.json','w') as f:
            json.dump(self.columns,f)
        
        print('Model Dumped')
                                                                                                                                     
        
        
```
